# K-Means Clustering for random dataset
K-means clustering is an unsupervised machine learning model (without labels). This clustering algoirthm searches for a preset number of clusters in an unlabled dataset.

The random dataset is created with sklearn's make_blobs function to create a dataset with isotropic Gaussian blobs.


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
```

## 1. Data Acquisition & Preparation

```python
from sklearn.datasets import make_blobs
data=make_blobs(n_samples=1000, n_features=2, centers=4, random_state=100, cluster_std=1.5)
# data=make_blobs(n_samples=1000, n_features=6, centers=4, random_state=100, cluster_std=2)
```

```python
data
```

### Exploratory Data Analysis


```python
#visualize the dataset with scatter plot
plt.scatter( data[0][:,0], data[0][:,1]) 
plt.show()
```

```python
#visualize the dataset with scatter plot with labels
plt.scatter( data[0][:,0], data[0][:,1], c=data[1]) 
plt.show()
```

With this scatter plot, we can see that this dataset has 4 clusters instead of 3. 


## 2. Model Selection
The dataset has 4 clusters and therefore we shall use K-means clustering as the ML model with no of clusters as 4.

## 3. Model Building and Training

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4,n_init=15)

#More hyperparameters can be input here
```

```python
# #standardization
# from sklearn.preprocessing import StandardScaler
# scl = StandardScaler()
# dfnorm = scl.fit_transform(data[0])
# dfnorm = pd.DataFrame(dfnorm)
```

```python
# kmeans.fit(dfnorm)
kmeans.fit(data[0])
```

## 4. Model Prediction

```python
#find where is the centre of clusters located at, the centroids
kmeans.cluster_centers_
```

```python
#predicted results
pred_results= kmeans.labels_
pred_results
```

```python
#actual results
data[1]
```

## 5. Model Evaluation

```python
#loss functions
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(data[1], pred_results)
```

The accuracy score is poor because we haven't mapped the existing categories with the categories created by the k-means model
